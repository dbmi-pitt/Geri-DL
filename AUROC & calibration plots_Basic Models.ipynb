```python
import csv
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, f1_score, accuracy_score, precision_recall_curve
```

### AUROC plots for Lotisgic Regression

```python


with open('LR_exp_1.csv', 'r') as readFile:
    reader = csv.reader(readFile)
    pr1 = list(reader)

with open('LR_exp_2.csv', 'r') as readFile:
    reader = csv.reader(readFile)
    pr2 = list(reader)

with open('LR_exp_3.csv', 'r') as readFile:
    reader = csv.reader(readFile)
    pr3 = list(reader)
    
fpr1 = [float(i) for i in pr1[0]]
tpr1 = [float(i) for i in pr1[1]]
fpr2 = [float(i) for i in pr2[0]]
tpr2 = [float(i) for i in pr2[1]]
fpr3 = [float(i) for i in pr3[0]]
tpr3 = [float(i) for i in pr3[1]]

# tpr(recall), fpr(1-specificity)
point1x = float(pr1[2][1])
point1y = float(pr1[2][0])
point2x = float(pr2[2][1])
point2y = float(pr2[2][0])
point3x = float(pr3[2][1])
point3y = float(pr3[2][0])

plt.figure(figsize=(6,4.5))
plt.plot(fpr1, tpr1, color='b', linewidth=1.2, label='LR_Exp. 1 (AUROC=%0.3f)' % auc(fpr1, tpr1), linestyle='-')
plt.plot(fpr2, tpr2, color='r', linewidth=1.2, label='LR_Exp. 2 (AUROC=%0.3f)' % auc(fpr2, tpr2), linestyle='-')
plt.plot(fpr3, tpr3, color='g', linewidth=1.2, label='LR_Exp. 3 (AUROC=%0.3f)' % auc(fpr3, tpr3), linestyle='-')
plt.plot([0, 1], [0, 1], color='darkorange', linestyle='--')
plt.scatter(point1x, point1y, s=100, c='b', marker='o', label='LR Exp. 1 Best F1 score')
plt.scatter(point2x, point2y, s=100, c='r', marker='o', label='LR Exp. 2 Best F1 score')
plt.scatter(point3x, point3y, s=100, c='g', marker='o', label='LR Exp. 3 Best F1 score')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.grid(linestyle=':')
plt.show()
```
